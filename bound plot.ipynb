```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nojump_data=np.load('data/nojump_data.npz')
jump_data=np.load('data/jump_data.npz')
b_grid=np.load('data/setup.npy')
```

```python
plt.plot(b_grid[:jump_data['end_point']],jump_data['lower_bound'][:jump_data['end_point']],label='lower bound with jump')
plt.plot(b_grid[:jump_data['end_point']],jump_data['upper_bound'][:jump_data['end_point']],label='upper bound with jump')
plt.plot(b_grid[:nojump_data['end_point']],nojump_data['lower_bound'][:jump_data['end_point']],label='lower bound without jump')
plt.plot(b_grid[:nojump_data['end_point']],nojump_data['upper_bound'][:jump_data['end_point']],label='upper bound without jump')
plt.legend()
plt.show()
```
